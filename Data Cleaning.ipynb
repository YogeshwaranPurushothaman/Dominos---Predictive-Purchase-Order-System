```python
df_sale['pizza_name_size_temp']=df_sale['pizza_name']+"_"+df_sale['pizza_size']
df_sale
```

```python
df_no_null=df_sale.dropna()
df_no_null
```

```python
# Create a mapping DataFrame from df_no_null
lookup_mapping = df_no_null.set_index('pizza_name_size_temp')['pizza_name_id'].to_dict()

# Fill null values in df_sale['pizza_name_id'] using the lookup mapping
df_sale['pizza_name_id'] = df_sale['pizza_name_id'].fillna(df_sale['pizza_name_size_temp'].map(lookup_mapping))

# Display the updated DataFrame
print(df_sale)
```

```python
# Update null values in totalprice by multiplying quantity and unit price
df_sale['total_price'] = df_sale['total_price'].fillna(df_sale['quantity'] * df_sale['unit_price'])

print("\nDataFrame After Updating Total Price:")
print(df_sale)
df_sale.info()
```

```python
df_sale['pizza_name_category_temp']=df_sale['pizza_name']+"_"+df_sale['pizza_category']
df_sale
```

```python
# Create a mapping DataFrame from df_no_null
lookup_mapping = df_no_null.set_index('pizza_name_category_temp')['pizza_ingredients'].to_dict()

# Fill null values in df_sale['pizza_ingredients'] using the lookup mapping
df_sale['pizza_ingredients'] = df_sale['pizza_ingredients'].fillna(df_sale['pizza_name_category_temp'].map(lookup_mapping))

# Display the updated DataFrame
print(df_sale.head())
```

```python
# Create a mapping dictionary from df_no_null
pizza_name_map = df_no_null.set_index('pizza_name_id')['pizza_name'].to_dict()

# Fill missing values in 'pizza_category' based on matching 'pizza_name'
df_sale['pizza_name'] = df_sale['pizza_name'].fillna(df_sale['pizza_name_id'].map(pizza_name_map))
df_sale.info()
```

```python
# Convert the 'order_date' column to datetime
df_sale['order_date'] = pd.to_datetime(df_sale['order_date'], errors='coerce')

# Create a new column 'day_of_week' based on 'order_date'
df_sale['day_of_week'] = df_sale['order_date'].dt.day_name()

# Create a new column 'day_type' to classify as 'Weekday' or 'Weekend'
df_sale['day_type'] = df_sale['day_of_week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')

# Display the updated DataFrame
print(df_sale)
```
